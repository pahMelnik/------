# BOOLEAN
##### csv - csv, где BOOLEAN выглядит как true/false
  - `csv1` - без конвертации колонок
  - `csv2` - c конвертацией колонок
##### om_csv - csv, где BOOLEAN выглядит как 1/0
  - `om_csv 1` - без конвертации колонок
  - `om_csv2` - c конвертацией колонок

```python
csv1 = file_to_dataframe("tests/test_data/csv.csv",
                  "csv",
                  invert_columns=False,
                  index_column='Days')
csv2 = file_to_dataframe("tests/test_data/csv.csv",
                  "csv",
                  invert_columns=False,
                  index_column='Days'
                  , columns_to_convert={'bool':['BOOLEAN'],
                                        'num': ['FLOAT', 'INT']})
om_csv1 = file_to_dataframe("tests/test_data/om_csv.csv",
                  "om_csv",
                  invert_columns=False,
                  index_column='Days',
                  encoding= 'windows 1251')
om_csv2 = file_to_dataframe("tests/test_data/om_csv.csv",
                  "om_csv",
                  invert_columns=False,
                  index_column='Days',
                  columns_to_convert={'bool':['BOOLEAN'],
                                      'num': ['FLOAT', 'INT']},
                  encoding= 'windows 1251')
csv1_bool_sr = csv1['BOOLEAN']
csv2_bool_sr = csv2['BOOLEAN']
om_csv1_bool_sr = om_csv1['BOOLEAN']
om_csv2_bool_sr = om_csv2['BOOLEAN']

csv1_int_sr = csv1['INT']
csv2_int_sr = csv2['INT']
om_csv1_int_sr = om_csv1['INT']
om_csv2_int_sr = om_csv2['INT']

csv1_float_sr = csv1['FLOAT']
csv2_float_sr = csv2['FLOAT']
om_csv1_float_sr = om_csv1['FLOAT']
om_csv2_float_sr = om_csv2['FLOAT']
```

```python
test_result = pd.DataFrame(data= {'type': ['bool', 'bool', 'num', 'num','float', 'float'],
                                  'csv1': [csv1_bool_sr[0], csv1_bool_sr.dtype, csv1_int_sr[0], csv1_int_sr.dtype, csv1_float_sr[0], csv1_float_sr.dtype],
                                  'csv2': [csv2_bool_sr[0], csv2_bool_sr.dtype, csv2_int_sr[0], csv2_int_sr.dtype, csv2_float_sr[0], csv2_float_sr.dtype],
                                  'om_csv1': [om_csv1_bool_sr[0], om_csv1_bool_sr.dtype, om_csv1_int_sr[0], om_csv1_int_sr.dtype, om_csv1_float_sr[0], om_csv1_float_sr.dtype],
                                  'om_csv2': [om_csv2_bool_sr[0], om_csv2_bool_sr.dtype, om_csv2_int_sr[0], om_csv2_int_sr.dtype, om_csv2_float_sr[0], om_csv2_float_sr.dtype]
                                  },
                           index= ['first_value', 'dtype', 'first_value', 'dtype', 'first_value', 'dtype'])
test_result
```
